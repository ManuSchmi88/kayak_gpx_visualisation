```python
import folium
import os
import gpxpy
from pandas import DataFrame
from geographiclib.geodesic import Geodesic
import mplleaflet
import matplotlib.pyplot as plt
```

```python
tue_geojson = os.path.join('data','tuebingen.geojson')
gpx_data_path = os.path.join('data', 'garmin_activity.gpx')
gpx_data = gpxpy.parse(open(gpx_data_path))
```

```python
track = gpx_data.tracks[0]
segment = track.segments[0]
```

```python
data = []
segment_length = segment.length_3d()
```

```python
for point_idx, point in enumerate(segment.points):
    data.append(
        [
            point.longitude,
            point.latitude,
            point.elevation,
            point.time,
            segment.get_speed(point_idx)
        ])
```

```python
columns = ['Longitude', 'Latitude', 'Altitude', 'Time', 'Speed']
df = DataFrame(data, columns = columns)
df.head()
```

```python
fig, ax = plt.subplots()
df = df.dropna()

ax.plot(
    df['Longitude'],
    df['Latitude'],
    color = 'darkorange',
    linewidth = 5, 
    alpha = 0.5)
gj = mplleaflet.fig_to_geojson(fig=fig)
```

```python
map_object = folium.Map(location=[48.512729,9.059229], zoom_start = 15, tiles = 'Stamen Terrain')
line = gj['features'][0]
gjson = folium.features.GeoJson(line)
map_object.add_child(gjson)
map_object.save('test.html')
```
